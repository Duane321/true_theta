## Running the LinUCB Algorithm

This notebook is to be paired with the blog post: https://truetheta.io/concepts/reinforcement-learning/lin-ucb/

```python
import numpy as np
import pandas as pd
import altair as alt

from lin_ucb_model import LinUCBSimulatorSimple
```

### Initialize the model.

```python
lin_ucb = LinUCBSimulatorSimple(
    T=200,
    error_std_k1=0.12,
    error_std_k2=0.22,
    alpha=.5,
    theta_k1=0.28,
    theta_k2=-0.14,
    intercept_k1=0.03,
    intercept_k2=0.02,
    lambda_ = 1
)
```

```python
performance_df = lin_ucb.run_many(sims=5000, re_simulate_truth=True)
performance_df
```

```python
alt.Chart(performance_df).mark_line(color='teal', strokeWidth=4).encode(x='t', y=alt.Y('mean_payoff', title='Expected Payoff')).properties(width=400, height=200)
```

```python
alt.Chart(performance_df).mark_line(strokeWidth=4, opacity=1, color='teal').encode(x='t', y=alt.Y('regret', title='Expected Regret')).properties(width=400, height=200, title='Expected Regret for LinUCB')
```
